# Phase and Bloch Sphere

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
import math
```

```python
qc = QuantumCircuit(2)
qc.draw(output='mpl')
```

```python
def run_on_simulators(circuit):
    circuit_sv = circuit.copy()
    circuit_sv.save_statevector()

    statevec_simulator = AerSimulator(method='statevector')
    statevec_job = statevec_simulator.run(circuit_sv)
    result = statevec_job.result()
    statevec = result.data(0)['statevector']

    circuit_qasm = circuit.copy()
    circuit_qasm.measure_all()

    qasm_simulator = AerSimulator()
    qasm_job = qasm_simulator.run(circuit_qasm, shots=1024)
    qasm_result = qasm_job.result()
    counts = qasm_result.get_counts()

    return statevec, counts
```

```python
statevec, counts = run_on_simulators(qc)
plot_bloch_multivector(statevec)
```

```python
plot_histogram(counts)
```

```python
qc.h(0)
qc.draw('mpl')
```

```python
statevec, counts = run_on_simulators(qc)
plot_bloch_multivector(statevec)
```

```python
plot_histogram(counts)
```

```python
qc.cx(0, 1)
qc.draw('mpl')
```

```python
statevec, counts = run_on_simulators(qc)
plot_bloch_multivector(statevec)
```

```python
plot_histogram(counts)
```

```python
qc2 = QuantumCircuit(2)
qc2.rx(math.pi / 4, 0) # Rotation about x-axis
qc2.ry(math.pi / 2, 1) # Rotation about y-axis
qc2.draw('mpl')
```

```python
statevec, counts = run_on_simulators(qc2)
plot_bloch_multivector(statevec)
```
