# Superdense Coding

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
```

```python
def run_on_simulator(circuit):
    circuit_qasm = circuit.copy()
    circuit_qasm.measure_all()

    qasm_simulator = AerSimulator()
    qasm_job = qasm_simulator.run(circuit_qasm, shots=1024)
    qasm_result = qasm_job.result()
    counts = qasm_result.get_counts()

    return counts
```

```python
a = 0 # Qubit 0 controlled by Alice
b = 1 # Qubit 1 controlled by Bob
```

## Alice applies I gate and sends the message to Bob |00>
Applying I gate will give the same 00 state.

```python
qc = QuantumCircuit(2)
qc.h(a)
qc.cx(a, b)
qc.id(a) # Alice applies I gate to her qubit
qc.cx(a, b)
qc.h(a)
qc.draw(output='mpl')
```

```python
counts = run_on_simulator(qc)
plot_histogram(counts)
```

## Alice applies X gate and sends the message to Bob |10>
Applying X gate will collapse the state 00 to 10.

```python
qc = QuantumCircuit(2)
qc.h(a)
qc.cx(a, b)
qc.x(a) # Alice applies X gate to her qubit
qc.cx(a, b)
qc.h(a)
qc.draw(output='mpl')
```

```python
counts = run_on_simulator(qc)
plot_histogram(counts)
```

## Alice applies Z gate and sends the message to Bob |01>
Applying Z gate will collapse the state 00 to 01.

```python
qc = QuantumCircuit(2)
qc.h(a)
qc.cx(a, b)
qc.z(a) # Alice applies Z gate to her qubit
qc.cx(a, b)
qc.h(a)
qc.draw(output='mpl')
```

```python
counts = run_on_simulator(qc)
plot_histogram(counts)
```

## Alice applies X and Z gate and sends the message to Bob |11>
Applying X gate will collapse the state 00 to 11.

```python
qc = QuantumCircuit(2)
qc.h(a)
qc.cx(a, b)
qc.x(a) # Alice applies X gate to her qubit
qc.z(a) # Alice applies Z gate to her qubit
qc.cx(a, b)
qc.h(a)
qc.draw(output='mpl')
```

```python
counts = run_on_simulator(qc)
plot_histogram(counts)
```
