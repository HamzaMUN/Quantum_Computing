# Classical Factorization Algorithm

```python
import matplotlib.pyplot as plt
from math import gcd
```

```python
N = 799
m = 8
```

```python
gcd_mN = gcd(m, N)
print(f'gcd(m, N) = gcd({m}, {N}) = {gcd_mN}')
```

```python
a = list(range(N))
r = [m**_ % N for _ in a]

plt.plot(a, r)
plt.xlabel('a')
plt.ylabel(f'{m}^P (mod {N})')
plt.show()
```

```python
P = a[r[1:].index(1) + 1]
print(f'Period: {P}')
```

```python
if P % 2 == 0:
    half_P = int(P / 2)
    mod_0 = (m**half_P - 1) * (m**half_P + 1) % N
    if mod_0 == 0:
        p = gcd(m**half_P + 1, N)
        p_mod_N = p % N
        if p_mod_N != 0:
            q = gcd(m**half_P - 1, N)
            print(f'p = {p} and q = {q}, where N = pq = {p} x {q} = {N}')
        else:
            print('m^P/2 + 1 = 0 mod N: Go back to step 1 and choose a different m.')
else:
    print('P is odd: Go back to step 1 and choose a differen m.')
```
