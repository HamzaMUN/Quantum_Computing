# Deutsch Algorithm

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
```

```python
# Two qubits quantum circuit |01>
qc = QuantumCircuit(2, 1)
qc.x(1) # Initializing q_1 as 1
qc.barrier()
qc.h(0)
qc.h(1)
qc.barrier()
qc.cx(0, 1) # Using CNOT as Oracle which is a balanced function
qc.barrier()
qc.h(0)
qc.barrier()
qc.measure(0, 0)
qc.draw(output='mpl')
```

```python
# Execute code on a simulator
backend = AerSimulator()
result = backend.run(qc, shots=1024).result()
counts = result.get_counts()
plot_histogram(counts)
```

As all the results are 1, the function (CNOT) is a **balanced** function.
